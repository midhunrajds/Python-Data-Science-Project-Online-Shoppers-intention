```python

x = df.iloc[:, [1, 6]].values

# checking the shape of the dataset
x.shape


from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
              init = 'k-means++',
              max_iter = 300,
              n_init = 10,
              random_state = 0,
              algorithm = 'full',
              tol = 0.001)
    km.fit(x)
    labels = km.labels_
    wcss.append(km.inertia_)    
plt.rcParams['figure.figsize'] = (10, 5)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters', fontsize = 15)
plt.ylabel('WCSS', fontsize = 15)
```

```python
newdf = pd.DataFrame({
                      "Admin_Dur": df['Administrative_Duration'],
                      "BounceRates": df['BounceRates']
                           })
```

```python
from sklearn.cluster import KMeans
clusters = KMeans(3) 
clusters.fit(newdf)
```

```python
newdf["clusterid"] = clusters.labels_
newdf[0:7]
```

```python

markers = ['+','^','.']
# We use different markers for the customer segments. 

sns.lmplot( x="Admin_Dur", y="BounceRates",
            data=newdf, hue = "clusterid", fit_reg=False, 
            markers = markers, height = 7, legend = False);

clusters.cluster_centers_
plt.legend(title='Clusterid', loc='upper right', labels=['Not Interested', 'Target', 'General'], fontsize = 15)
plt.title("Clustering based on Administrative Duration")

```
